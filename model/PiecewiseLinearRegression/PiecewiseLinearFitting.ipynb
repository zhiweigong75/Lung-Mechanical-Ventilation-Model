```python
import numpy as np
import matplotlib.pyplot as plt
import pwlf
import pandas as pd
import time
```

```python
# import PEEP7_RR12_CT0.01_PaO25
data = []
volume = []
flow = []
pressure = []
Time = []

data1 = pd.read_excel('PEEP7_RR12_CT0.01_PaO25.xlsx')
volume1 = data1['Volume']
flow1 = data1['Flow']
pressure1 = data1['Pressure']
time1 = data1['Time']
data.append(data1)
volume.append(volume1)
flow.append(flow1)
pressure.append(pressure1)
Time.append(time1)

# import PEEP7_RR13_CT0.01_PaO25
data2 = pd.read_excel('PEEP7_RR13_CT0.01_PaO25.xlsx')
volume2 = data2['Volume']
flow2 = data2['Flow']
pressure2 = data2['Pressure']
time2 = data2['Time']
data.append(data2)
volume.append(volume2)
flow.append(flow2)
pressure.append(pressure2)
Time.append(time2)

# import PEEP7_RR14_CT0.01_PaO25
data3 = pd.read_excel('PEEP7_RR14_CT0.01_PaO25.xlsx')
volume3 = data3['Volume']
flow3 = data3['Flow']
pressure3 = data3['Pressure']
time3 = data3['Time']
data.append(data3)
volume.append(volume3)
flow.append(flow3)
pressure.append(pressure3)
Time.append(time3)

# import PEEP7_RR15_CT0.01_PaO25
data4 = pd.read_excel('PEEP7_RR15_CT0.01_PaO25.xlsx')
volume4 = data4['Volume']
flow4 = data4['Flow']
pressure4 = data4['Pressure']
time4 = data4['Time']
data.append(data4)
volume.append(volume4)
flow.append(flow4)
pressure.append(pressure4)
Time.append(time4)

# import PEEP7_RR16_CT0.01_PaO25
data5 = pd.read_excel('PEEP7_RR16_CT0.01_PaO25.xlsx')
volume5 = data5['Volume']
flow5 = data5['Flow']
pressure5 = data5['Pressure']
time5 = data5['Time']
data.append(data5)
volume.append(volume5)
flow.append(flow5)
pressure.append(pressure5)
Time.append(time5)
```

## Pressure fitting

```python
# PEEP7_RR12
# First repetition
y1 = pressure1[0:501]
x1 = time1[0:501]
my_pwlf = pwlf.PiecewiseLinFit(x1, y1)
res = my_pwlf.fit(4)
r_square = my_pwlf.r_squared()
# predict for the determined points
xHat = x1
yHat = my_pwlf.predict(xHat)

breakpoints = my_pwlf.fit_breaks
print('breakpoints location (time) are :', breakpoints)
print('R^2 value is:', r_square)

# plot the results
plt.figure()
plt.plot(x1, y1, 'o')
plt.plot(xHat, yHat, '-')
plt.show()
```

```python
# PEEP7_RR12
# whole repetition
y1 = flow1
x1 = time1
my_pwlf = pwlf.PiecewiseLinFit(x1, y1, degree = 2)
res = my_pwlf.fit(12)
r_square = my_pwlf.r_squared()
# predict for the determined points
xHat = x1
yHat = my_pwlf.predict(xHat)

breakpoints = my_pwlf.fit_breaks
print('breakpoints location (time) are :', breakpoints)
print('R^2 value is:', r_square)

# plot the results
plt.figure()
plt.plot(x1, y1, 'o')
plt.plot(xHat, yHat, '-')
plt.show()

# run 31 minutes
```

```python
# Running 5 datasets for flow fitting

for i in range(5):
    # whole repetition
    start = time.time()
    
    y1 = flow[i]
    x1 = Time[i]
    my_pwlf = pwlf.PiecewiseLinFit(x1, y1, degree = 2)
    if i == 0:
        res = my_pwlf.fit(12)
    elif i == 1:
        res = my_pwlf.fit(15)
    else:
        res = my_pwlf.fit(18)
    r_square = my_pwlf.r_squared()
    # predict for the determined points
    xHat = x1
    yHat = my_pwlf.predict(xHat)

    breakpoints = my_pwlf.fit_breaks
    print('breakpoints location (time) are :', breakpoints)
    print('R^2 value is:', r_square)

    # plot the results
    plt.figure()
    plt.plot(x1, y1, 'o')
    plt.plot(xHat, yHat, '-')
    plt.show()
    end = time.time()
    print(end - start)
```
